### Associate Nuclear coordinates with closest Branch

```python
# coordinates = list of (y, x) points to check (e.g., nuclei coordinates) detected with local_maximum_peak
coordinates_filtered = []
label_ids = []
thrown_out = []

# Associate nuclei with the closest branch in the branch img with a distance cutoff
coords, branch_ids, branch_tree = M.build_branch_kdtree(labels)

for y, x in coordinates:
    y,x = int(y), int(x)
    label_id = labels[y, x]
    
    if label_id > 0:  # it's already a branch
        coordinates_filtered.append((y, x))
        label_ids.append(label_id)
    else: 
        closest_label_id = M.find_closest_branch(y, x, branch_tree, branch_ids, max_dist=10)
        if closest_label_id is not None:
            coordinates_filtered.append((y, x))
            label_ids.append(closest_label_id)  # tag as branch
        else: 
            thrown_out.append((y, x))
            
# count how many times a nuclear blob is detected in the network
label_counts = Counter(label_ids)  # {branch_id: count}

# Add a new column to branch_df
branch_df["n_nuclei"] = branch_df["branch_id"].map(lambda b: label_counts.get(b, 0))
```

```python
# Inspect coordinates in Napari
nuclei_labels = io.imread('nuclei_labels.tiff')
viewer.add_labels(nuclei_labels)
viewer.add_points(coordinates_filtered, name='detected nuclei')
```
