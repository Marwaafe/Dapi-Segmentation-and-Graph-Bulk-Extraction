```python
import napari 
from glob import glob
from pathlib import Path
from skimage import io
import numpy as np
from skimage.exposure import match_histograms

def add_dims_to_tips(tips):
    arrs = []
    for index in range(len(tips)):
        arr = np.load(tips[index])
        extra_column = np.full((arr.shape[0],1), index)
        new_arr = np.hstack([extra_column, arr])
        arrs.append(new_arr)
    final_array = np.concatenate(arrs,axis=0)    
    return final_array
tips_array= add_dims_to_tips(tips)
tips_array.shape

def create_stack(paths):
    imgs = [io.imread(path) for path in paths]
    max_y = np.max([img.shape[0] for img in imgs])
    max_x = np.max([img.shape[1] for img in imgs])
    reference = io.imread(paths[-1])
    
    stack = np.empty((len(imgs),max_y,max_x))
    for i,img in enumerate(imgs):
        diff_y =  max_y - img.shape[0] 
        diff_x =  max_x - img.shape[1]
        matched = match_histograms(img,reference)
        stack[i,diff_y:,diff_x:] = matched
        
    return stack
```

```python
dirs = glob(fr'S:\testing_chip_analysis\processed_improved\*')
dirs = [Path(_dir).name for _dir in dirs]
dirs
```

```python
i = 1
chip = dirs[i]
path_branches = fr'S:\testing_chip_analysis\processed_improved\{chip}\branch\*.tiff'
path_tips = fr'S:\testing_chip_analysis\processed_improved\{chip}\tips\*.npy'
path_bfs = fr'S:\testing_chip_analysis\processed_improved\{chip}\resized\*.tif'
branches = glob(path_branches)
bfs = glob(path_bfs)
tips = glob(path_tips)
bf_img = create_stack(bfs)
branch_img = np.concatenate([np.expand_dims(io.imread(branch),0) for branch in branches])
```

```python
tips_array= add_dims_to_tips(tips)
```

```python
viewer = napari.Viewer()
```

```python
viewer.add_image(bf_img)
```

```python
viewer.add_labels(branch_img)
```

```python
viewer.add_points(tips_array, face_color='red', border_color='black', opacity=0.5)
```
