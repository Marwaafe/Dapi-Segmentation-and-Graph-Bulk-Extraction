```python
rng = np.random.default_rng(seed=42)  
def random_nonoverlapping_crops(image, labels, crop_size=(600, 600), n_crops=10, min_label_pixels=(600*600)/20, rng=None):
    if rng is None:
        rng = np.random.default_rng()

    H, W = labels.shape
    ch, cw = crop_size
    crops, crop_labels = [], []

    # mask of available space (True = available)
    available = np.ones((H, W), dtype=bool)

    attempts = 0
    while len(crops) < n_crops and attempts < n_crops * 20:
        attempts += 1

        # pick a random top-left corner
        y = rng.integers(0, H - ch + 1)
        x = rng.integers(0, W - cw + 1)

        # check if this crop is fully available
        if not available[y:y+ch, x:x+cw].all():
            continue

        img_crop = image[y:y+ch, x:x+cw]
        lbl_crop = labels[y:y+ch, x:x+cw]

        if np.count_nonzero(lbl_crop) >= min_label_pixels:
            crops.append(img_crop)
            crop_labels.append(lbl_crop)

            # mark this crop area as unavailable
            available[y:y+ch, x:x+cw] = False

    return crops, crop_labels
```

```python
from skimage.future import trainable_segmentation as future
from sklearn.ensemble import RandomForestClassifier

# assume napari layers alternate: image1, mask1, image2, mask2, ...
layers = viewer.layers

images = []
masks = []

for i in range(0, len(layers), 2):
    images.append(layers[i].data)
    masks.append(layers[i+1].data)
```

```python
def generate_feature_stack(img):
    """Generates feature stack for training data."""
    
    sigma_min = 1
    sigma_max = 16

    features_func = partial(feature.multiscale_basic_features,
                        intensity=True, edges=False, texture=True,
                        sigma_min=sigma_min, sigma_max=sigma_max,
                        channel_axis=0)

    features = features_func(np.expand_dims(img,axis=0))

    return features
def create_classifier_and_features(img, training_labels, features):
    """Fits classifier to training data and annotation."""
    clf = RandomForestClassifier(n_estimators=50, n_jobs=-1,
                             max_depth=10, max_samples=0.05)

    mask = training_labels > 0
    training_data = features[mask]
    training_labels_1D = training_labels[mask].ravel()
    clf.fit(training_data, training_labels_1D)

    return clf, features
def predict_probs(clf,features):
    """Generate probability map for the different class prediction of a pretrained classifier."""
    # reshape features to 1D shape to fit probabilities
    shape = features.shape
    if features.ndim > 2:
        features = features.reshape((-1, shape[-1]))
    predicted_probs = clf.predict_proba(features)

    # reshape predicted probabilities back to original features shape
    class_probs=[]
    for i in range(predicted_probs.shape[-1]):
        predicted_prob=predicted_probs[...,i]
        class_probs.append(np.expand_dims(predicted_prob.reshape(shape[:-1]),axis=-1))

    #return array with class probabilities with same shape as feature stack
    probabilities_output= np.concatenate(class_probs,axis=-1)
    
    return probabilities_output
```

```python
rng = np.random.default_rng(seed=43)  
new_crops, crops_labels = random_nonoverlapping_crops(DAPI, labels, rng=rng)
```

```python
def generate_training_data(img_list, label_list):
    """Generates combined feature stack and labels from multiple images."""
    all_features = []
    all_labels = []

    for img, labels in zip(img_list, label_list):
        features = generate_feature_stack(img)
        mask = labels > 0  # use only annotated pixels
        all_features.append(features[mask])
        all_labels.append(labels[mask].ravel())

    # Stack all features and labels across images
    all_features = np.vstack(all_features)
    all_labels = np.hstack(all_labels)

    return all_features, all_labels
```

```python
all_feautures, all_labels = generate_training_data(images,masks)
```

```python
def train_classifier_on_multiple_images(img_list, label_list):
    X_train, y_train = generate_training_data(img_list, label_list)

    clf = RandomForestClassifier(
        n_estimators=50,
        max_depth=10,
        max_samples=0.05,
        n_jobs=-1
    )
    clf.fit(X_train, y_train)

    return clf
```

```python
clf = train_classifier_on_multiple_images(images,masks)
```

```python
def predict_image_probs(clf, img):
    features = generate_feature_stack(img)
    probs = predict_probs(clf, features)
    return probs
```
