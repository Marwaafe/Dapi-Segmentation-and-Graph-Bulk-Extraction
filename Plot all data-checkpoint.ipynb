```python
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
import pandas as pd
from pathlib import Path
sns.set_theme(style="darkgrid")

# Set global font color to white
plt.rcParams.update({
    'text.color': 'white',       # General text color
    'axes.labelcolor': 'white',  # X and Y axis labels
    'xtick.color': 'white',      # X-axis tick labels
    'ytick.color': 'white',      # Y-axis tick labels
    'axes.edgecolor': 'white',   # Axes border color
    'axes.titlecolor': 'white',  # Title color
    'font.size': 14,             # General font size
    'axes.titlesize': 16,        # Title font size
    'axes.labelsize': 14,        # X and Y label font size
    'xtick.labelsize': 12,       # X-axis tick font size
    'ytick.labelsize': 12        # Y-axis tick font size
})

path = fr'S:\testing_chip_analysis\processed_improved\*\csv\*.csv'
files = glob(path)
```

```python
import functools

df_list = []
for file in files:
    chip_id, *extras = Path(file).stem.split('+')
    plant_hormone, plant_extract = (extras + [' + None', ' + None'])[:2]
    
    data = pd.read_csv(file)
    data['id'], data['plant_hormone'], data['plant_extract'] = chip_id, plant_hormone, plant_extract
    df_list.append(data)
    
merged_df = functools.reduce(lambda x, y: x.append(y, ignore_index=True), df_list)
merged_df.head(4)
```

```python
merged_df['all_branches_lengths_um'] = merged_df['all_branches_lengths_um'].str.strip('[').str.strip(']').str.split(',')
merged_df['tips_lengths_um'] = merged_df['tips_lengths_um'].str.split(';')
merged_df['other_lengths_um'] = merged_df['other_lengths_um'].str.split(';')
```

```python
exploded_df= merged_df.explode(['all_branches_lengths_um'])
exploded_df.all_branches_lengths_um = exploded_df.all_branches_lengths_um.astype(float)
```

```python
exploded_df['all_branches_lengths_mm']= exploded_df.all_branches_lengths_um/1000
```

```python
fig, ax = plt.subplots(figsize=(25,5), ncols=2)

sns.boxplot(data=exploded_df[exploded_df['strain']=='C2'],x='day',y='all_branches_lengths_mm', hue='label', hue_order =hue_order,ax=ax[0])
ax[0].legend().remove()
sns.boxplot(data=exploded_df[exploded_df['strain']=='A5'],x='day',y='all_branches_lengths_mm', hue='label', hue_order = hue_order,ax=ax[1])
ax[0].set_ylabel('Branch lengths (mM)')
ax[1].set_ylabel('Branch lengths (mM)')
ax[0].set_ylim(-10,300)
ax[0].set_title('C2; homokaryon')
ax[1].set_ylim(-10,300)
ax[1].set_title('A5; dikaryon')
ax[1].legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False)
fig.suptitle('Branch length distributions (mM) over time')
plt.savefig("branch length distributions over time.png", dpi=600, transparent=True, bbox_inches='tight')
```

```python
exploded_tips_df= merged_df.explode(['tips_lengths_um'])
exploded_tips_df['branch_lengths'] = exploded_tips_df.tips_lengths_um.astype(float)
exploded_tips_df['branch_type'] = 'tip'
exploded_other_df= merged_df.explode(['other_lengths_um'])
exploded_other_df['branch_lengths'] = exploded_other_df.other_lengths_um.astype(float)
exploded_other_df['branch_type'] = 'other'
```

```python
# Concatenate both dataframes (stack them vertically)
both_df = pd.concat([exploded_tips_df, exploded_other_df], ignore_index=True)
```

```python
fig, ax = plt.subplots(figsize=(25,5), ncols=2)

sns.stripplot(data=both_df[(both_df['strain']=='C2') & (both_df.label==sel_label)],x='day',y='branch_lengths', color='darkorange', hue_order =hue_order,ax=ax[0],dodge=True, s=3)
# plot the mean line
sns.boxplot(showmeans=True,
            meanline=True,
            meanprops={'color': 'red', 'ls': '-', 'lw': 2},
            medianprops={'visible': False},
            whiskerprops={'visible': False},
            zorder=10,
            x="day",
            y="branch_lengths",
            data=both_df[(both_df['strain']=='C2') & (both_df.label==sel_label)],
            showfliers=False,
            showbox=False,
            showcaps=False,
            ax=ax[0])
ax[0].legend().remove()
sns.stripplot(data=both_df[(both_df['strain']=='A5') & (both_df.label==sel_label)],x='day',y='branch_lengths', color='darkorange', hue_order = hue_order,ax=ax[1],dodge=True,s=2)
ax[0].set_ylabel('edge lengths (uM)')
ax[1].set_ylabel('edge lengths (uM)')
ax[0].set_ylim(-10,1000)
ax[0].set_title('')
ax[1].set_ylim(-10,1000)
ax[1].set_title('A5; dikaryon')
ax[1].legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False)
fig.suptitle('Branch length distributions (uM) over time')
plt.savefig("branch length distributions strip.png", dpi=600, transparent=True, bbox_inches='tight')
```
